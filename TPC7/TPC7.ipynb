#### a) Especifique uma função que dada uma string e uma substring não vazia, calcula  o número de vezes em que a substring aparece na string, sem que haja sobreposição de substrings:

```python
def strCount(s, subs):
    res = 0
    i = 0
    while i <= len(s) - len(subs):
        if s[i:i+len(subs)] == subs:
            res += 1
            i += len(subs)  
        else:
            i += 1         
    return res       


print(strCount("catcowcat", "cat")) # --> 2
print(strCount("catcowcat", "cow")) # --> 1
print(strCount("catcowcat", "dog")) # --> 0
```

#### b) Especifique uma função que recebe uma lista de números inteiros positivos e devolve o menor produto que for possível calcular multiplicando os 3 menores inteiros da lista:

```python
def produtoM3(lista):
    res = 1
    ordem = sorted(lista)
    for i in ordem[:3]:
        res = res * i 
    return (f"{res} = {ordem[0]} * {ordem[1]} * {ordem[2]}")


print(produtoM3([12,3,7,10,12,8,9]))
print(produtoM3([12,3,4,10,12,-1,9]))
```

#### c) Especifique uma função que dado um número inteiro positivo, repetidamente adiciona os seus dígitos até obter apenas um dígito que é retornado como resultado:

```python
def reduxInt(n):
    num = str(n) 
    res = 0
    for i in num:
        res = res + int(i)
        final = str(res) 
        som = 0
        for j in final:
            som = som + int(j)
    return som
        

print(reduxInt(38))
print(reduxInt(777))
print(reduxInt(823))
print(reduxInt(185312))
```

#### d) Especifique uma função que recebe duas strings, `string1` e `string2`, e devolve o índice da primeira ocorrência de `string2` em `string1`, caso não ocorra nenhuma vez a função deverá retornar `-1`:

```python
def myIndexOf(s1, s2):
    s1norma = s1.lower()
    s2norma = s2.lower()
    if s2norma in s1norma:
        return s1norma.index(s2norma)
    else:
        return -1

print(myIndexOf("Hoje está um belo dia de sol!", "belo"))
print(myIndexOf("Hoje está um belo dia de sol!", "chuva"))
print(myIndexOf("Hoje está um belo dia de sol!", "hoje"))
```

#### b)  `postsAutor`, que devolve a lista de posts de um determinado autor:

#### d) `insPost`, que acrescenta um novo post à rede social a partir dos parâmetros recebidos e devolve a nova rede social. 
    
O campo `id` devrá ser calculado a partir dos já existentes, por exemplo, se a rede tiver posts com id `p1`, `p2` e `p3`, o novo `id` deverá ser `p4`.

```python
def insPost(redeSocial, conteudo, autor, dataCriacao, comentarios, autorcom):

    if len(redeSocial) == 0:
        novoID= "p1"
    else:
        novoID = f"p{len(redeSocial)+1}"
    
    novoPost = {
        'id': novoID,
        'conteudo': conteudo,
        'autor': autor,
        'dataCriacao': dataCriacao,
        'comentarios': 
        [{'comentario' : comentarios, 'autor' : autorcom}]
        
    }
    redeSocial.append(novoPost)
    return redeSocial

insPost(MyFaceBook, "Que dia cansativo!", "Cláudia", "2025-11-02", "É a vida.","Ana")
insPost(MyFaceBook, "Hoje fui cortar o cabelo", "Tiago", "2025-11-08", "Fica-te bem o novo corte", "Joana")
```

#### e)  `remPost`, que remove um post da rede, correspondente ao `id` recebido.

```python
def remPost(redeSocial, id):
    for i in redeSocial:
        if i['id'] == id:
            redeSocial.remove(i)
    return redeSocial

remPost(MyFaceBook, "p5")

```

#### f) `postsPorAutor`, que devolve uma distribuição de posts por autor (à semelhança do que foi feito nas aulas).

```python
def postsPorAutor(redeSocial):
    distrib = {}
    for post in redeSocial:
        autor = post['autor']
        if autor in distrib:
            distrib[autor] += 1
        else:
            distrib[autor] = 1
    return distrib

print(postsPorAutor(MyFaceBook))
```

#### g) `comentadoPor`, que recebe um autor e devolve a lista de posts comentados por esse autor.

```python
def comentadoPor(redeSocial, autorcom):
    comentarios = []
    for post in redeSocial:
        for comentario in post['comentarios']:
            if comentario['autor'] == autorcom:
                comentarios.append(comentario['comentario'])
    return comentarios


print(comentadoPor(MyFaceBook, "jj"))
print(comentadoPor(MyFaceBook, "mfs"))
print(comentadoPor(MyFaceBook, "jcr"))
print(comentadoPor(MyFaceBook, "Ana"))
print(comentadoPor(MyFaceBook, "nini"))
    
```
