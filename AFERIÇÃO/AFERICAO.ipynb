```python
#1 a)
def contaEspacos(s):
    if s == "":
        esp = 0
    elif s[0] == " ":
        esp = 1 + contaEspacos(s[1:])
    else:
        esp = 0 + contaEspacos(s[1:])
    return esp

contaEspacos("Olá eu sou a Cláudia")
```

```python
#1 b)
def ordenada(lista):
    if len(lista) <= 1:
        res = True
    elif lista[0] <= lista[1]:
        res = ordenada(lista[1:])
    else: 
        res = False
    return res

print(ordenada([1,3,3,2,6]))
print(ordenada([1,2,3,4,6]))
```

```python
#1 c)

def existeElem(lista, elem):
    if lista == []:
        res = False
    elif elem == lista[0]:
        res = True
    else: 
        res = existeElem(lista[1:], elem)
    return res

print(existeElem([1,2,6,4,24],24))
print(existeElem([1,2,6,4,24],7))
```

```python
#2 a)
def maiorDivisor(N):
    res = {}
    for i in range(1, N+1):
        maior = 1
        for d in range(1, 10):
            if i % d == 0:
                maior = d
        res[i] = maior
    return res

print(maiorDivisor(20))
```

```python
#2 b)
def mdc(a, b):
    while b != 0:
        a, b = b, a % b
    return a

def mmc(a, b):
    return (a*b) // mdc(a, b)

mmc(12,16)
```

```python
#2 c)
def f2c(dict):
    res = {}
    for i in dict:
        fah = dict[i]
        cel = (fah-32)/1.8
        res[i] = (fah, cel)
    return res

fahrenheit = {'t1':-30, 't2':-20, 't3':-10, 't4':0}
f2c(fahrenheit)
```

```python
#2 d)
def reduxInt(n):
    while n >= 10:
        soma = 0
        for i in str(n):
            soma += int(i)
            n = soma
    return soma

print(reduxInt(998))
```

```python
#3 a)
def listar_unidades(hospital):
    res = []
    for i in hospital:
        res.append(i[0])
    return res

listar_unidades(hospital)
```

```python
#3 b)
def camas_disponiveis(hospital):
    res = 0
    for i in hospital:
        camas = i[2]
        for cama in camas:
            if cama[1] == False:
                res += 1
    return res

camas_disponiveis(hospital)
```

```python
#3 c)

def listas_medicos(hospital, unidade):
    res = []
    for i in hospital:
        if i[0] == unidade:
            res.append(i[1])
    return res

print(listas_medicos(hospital, "Unidade de Cardiologia"))
print(listas_medicos(hospital, "Unidade de Pediatria"))
print(listas_medicos(hospital, "Unidade de Dermatologia"))
```

```python
#3 d)

def existe_medico(hospital, unid, medico):
    res = False
    for i in hospital:
        if unid == i[0]:
            medicos = i[1]
            for med in medicos:
                if med[1] == medico:
                    res = True
    return res

print(existe_medico(hospital, "Unidade de Cardiologia", "Dra. Maria")) 
print(existe_medico(hospital, "Unidade de Cardiologia", "Dr. Alberto"))
print(existe_medico(hospital, "Unidade de Cardiologia", "Dr. João"))
```

```python
#3 e)
def adicionar_medico(hospital, unid, medico):
    for i in hospital:
        if i[0] == unid:
            medicos = i[1]
            existe = False
            for med in medicos:
                if med[1] == medico:
                    existe = True
            if not existe:
                medicos.append((len(medicos)+1, medico))
    return hospital

print(adicionar_medico(hospital, "Unidade de Cardiologia", "Dra. Paula"))
print(adicionar_medico(hospital, "Unidade de Pediatria", "Dr. Hugo"))
```

```python
#3 f)
def ocupar_cama(hospital, unidade, id):
    novo = []
    for i in hospital:
        if i[0] == unidade:
            novas_camas = []
            for cama in i[2]:
                if cama[0] == id and not cama[1]:
                    novas_camas.append((cama[0], True))
                else:
                    novas_camas.append(cama)
            novo.append((i[0], i[1], novas_camas))
        else:
            novo.append(i)
    return novo

print(ocupar_cama(hospital, "Unidade de Cardiologia", 101))
```

```python
#3 g)
#escolhi guardar num ficheiro txt
def guardar_hospital(hospital, nome):
    with open(nome, 'w', encoding='utf-8') as f:
        for unidade in hospital:
            f.write(f"{unidade[0]}\n")
            for medico in unidade[1]:
                f.write(f"{medico[0]} - {medico[1]}\n")
            for cama in unidade[2]:
                if cama[1] == True:
                    estado = "Ocupada" 
                else:
                    estado = "Livre"
                f.write(f"{cama[0]} - {estado}\n")
            f.write("\n")
            
guardar_hospital(hospital, "hospitall.txt")
```

```python
#3 i)
def cama_ocupada(hospital):
    ocupadas = 0
    total = 0
    for i in hospital:
        camas = i[2]
        total += len(camas)
        for cama in camas:
            if cama[1] == True:
                ocupadas += 1
    proporcao = (ocupadas/total)*100
    return proporcao

print(cama_ocupada(hospital))
```

```python
#3 j)
def unidades_ocupacao_critica(hospital):
    res = []
    for i in hospital:
        nome = i[0]
        camas = i[2]
        ocupadas = 0
        for cama in camas:
            if cama[1]:
                ocupadas += 1
        proporcao = (ocupadas / len(camas)) * 100
        if proporcao >= 80:
            res.append((nome, proporcao))
    return res

print(unidades_ocupacao_critica(hospital))
```
